```python
ann_stats = pd.read_csv("Seasons_Stats.csv")
ann_stats.head()
```

```python
TOT_list = ann_stats[ann_stats["Tm"] == 'TOT'].index
TOT_list
```

```python
ann_without_TOT = ann_stats.drop(TOT_list)
ann_without_TOT
```

```python
ann_cleaned = ann_without_TOT[ann_without_TOT["Player"].notna()]
ann_cleaned
```

```python
ann_cleaned["last letter"] = [ann_cleaned["Player"][i][-1:]=='*' for i in ann_cleaned.index]
hof = ann_cleaned[ann_cleaned["last letter"] == True].drop("last letter", axis=1)
hof
```

```python
agg_hof = hof.groupby(["Player"]).agg('sum')[["G", "GS", "MP", "TRB", "AST", "STL", "BLK", "TOV", "PF", "PTS"]]
agg_hof
```

```python
agg_hof.sort_values("PTS", ascending=False).head(20)
```
